## Data Encoding :- 
`Converting the categorical values into numerical values so that the model is able to understand and differentiate`

#### Ways to do it :- 
1. Nominal or one Hot Encoding
2. Label and Ordinal Encoding
3. Target Guided Ordinal Encoding


****************************************************************************

### Nominal / One-Hot Encoding
`Definition`: A technique to convert categorical (nominal) data into a numerical format for Machine Learning models. It creates a binary column for every unique category in a feature.

How it Works: If a feature "Color" has 3 categories: Red, Blue, Green:

Red becomes [1, 0, 0]

Blue becomes [0, 1, 0]

Green becomes [0, 0, 1]

### Disadvantages
`Sparse Matrix`: Since only one value is "1" and the rest are "0," the resulting table is mostly empty (zeros), which is memory-inefficient.

`Feature Explosion`: If a column has high cardinality (e.g., 500 unique Cities), One-Hot Encoding adds 500 new features, leading to the Curse of Dimensionality.

`Multicollinearity (Dummy Variable Trap)`: One feature can be predicted from the others (e.g., if it’s not Red or Blue, it must be Green).

Fix: Use drop_first=True to remove one redundant column.


#### When to Use
> When categories have no natural order (e.g., Countries, Colors).

> When the number of unique categories is low (typically < 10-15).

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
## Create a Simple dataFrame
df = pd.DataFrame({
    'Color' : ['Red' , 'Green' , 'Blue' , 'Blue' , 'Green' , 'Red']
})

df
```

```python
## Create an instance of OneHotEncoder 
encoder = OneHotEncoder()
```

```python
color_encoded = encoder.fit_transform(df[['Color']]).toarray()
```

### columns = encoder.get_feature_names_out()
The method looks at the original column name and the unique categories inside it to create a list of strings.

Original Data: A column named Color with categories Red, Green, and Blue.

Method Output: A list: ['Color_Red', 'Color_Green', 'Color_Blue'].

```python
encoder_df = pd.DataFrame(color_encoded , columns = encoder.get_feature_names_out())
```

```python
encoder_df
```

```python
## For new data 
encoder.transform([['Blue']]).toarray()
```

```python
pd.concat([df , encoder_df] , axis = 1)
```

# Label Encoding :- 
 `Definition`: A technique used to convert categorical text labels into numerical values. Each unique category is assigned an integer (e.g., 0, 1, 2) based on alphabetical order.

```python
from sklearn.preprocessing import LabelEncoder 
lbl_encoder = LabelEncoder()
```

## Ordinal Encoding 
It is used to encode categorical data that have an intrinsic order or ranking. In this technique, each category is assigned a numerical value based on its position in the order. For example, if we have a categorical variable "education level" with four possible values (high school, college, graduate, post-graduate), we can represent it using ordinal encoding as follows:

1. High school: 1
2. College: 2
3. Graduate: 3
4. Post-graduate: 4



```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
df = pd.DataFrame({
    'Size' : ['Small' , 'Large' , 'Small' , 'Medium' , 'Large']
})
```

```python
df
```

```python
encoder = OrdinalEncoder(categories = [['Small' , 'Medium' , 'Large']])
```

```python
encoder.fit_transform(df[['Size']])
```

```python
encoder.transform([['Small']])
```

## Target Guided Ordinal Encoding 

```python
data = {
    'City': ['Mumbai', 'Delhi', 'Bangalore', 'Mumbai', 'Bangalore', 'Delhi', 'Mumbai', 'Chennai'],
    'Price': [12000, 15000, 18000, 12500, 17500, 14800, 12200, 11000]
}
```

```python
df = pd.DataFrame(data)
df
```

```python
mean_price = df.groupby('City')['Price'].mean().to_dict()
mean_price
```

```python
df['City_Encoded'] = df['City'].map(mean_price)
```

```python
df
```

```python
df[['City' , 'City_Encoded']]
```
