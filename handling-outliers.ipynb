## Outliers :- A data point or observation that behaves very differently in comparision to other data points. 


```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
lst_marks = [45,32,56,75,89,54,32,89,90,87,67,54,45,98,99,67,74]
minimum , q1 , median , q3 , maximum = np.quantile(lst_marks , 
                                                     [0,0.25,0.50,0.75,1.0])
```

```python
minimum , q1 , median , q3 , maximum 
```

## IQR (interquartile range) 
its the distance between the 75th percentile and the 25th percentile 
then i will make lower_fence and upper_fence anything above higher_fence or anything lower than lower_fence will be considered as an outlier|


```python
IQR = q3 - q1
IQR
```

```python
lower_fence = q1 - 1.5*(IQR) 
higher_fence = q3 + 1.5*(IQR)
```

```python
lower_fence , higher_fence
```

## WE can also know about outliers using boxplot 

```python
sns.boxplot(lst_marks)
```
